# 1. Two Sum


```python
def  TwoSum(nums,target):
  seen={}
  n=len(nums)
  for i,num in enumerate(nums):
    if (target-num) in seen:
      return [seen[target-num],i]
    seen[num]=i

  return -1

arr=[2,4,5,3,7]
target=9
print(TwoSum(arr,target))
```

# 2. Two Sum Sorted Array

```python
def TwoSumSorted(nums,target):
  right=len(nums)-1
  left=0
  for i in range(len(nums)):
    current_sum=nums[left]+nums[right]
    if current_sum==target:
      return [left,right]
    elif current_sum>target:
      right-=1
    else:
      left+=1
  return -1

arr=[2,4,5,7,9]
target=9
print(TwoSumSorted(arr,target))
```

# 3. Three Sum

```python
def threeSumOptimal(nums, target):
    nums.sort()  # Step 1: Sort the array
    n = len(nums)
    result = []

    for i in range(n - 2):  # Step 2: Fix first element
        if i > 0 and nums[i] == nums[i - 1]:
            continue  # Skip duplicate values for i

        left = i + 1
        right = n - 1

        while left < right:
            curr_sum = nums[i] + nums[left] + nums[right]
            if curr_sum == target:
                result.append([nums[i], nums[left], nums[right]])

                # Skip duplicates for left and right
                while left < right and nums[left] == nums[left + 1]:
                    left += 1
                while left < right and nums[right] == nums[right - 1]:
                    right -= 1

                left += 1
                right -= 1

            elif curr_sum < target:
                left += 1
            else:
                right -= 1

    return result if result else -1


arr = [-1, 0, 1, 2, -1, -4]
target = 0
print(threeSumOptimal(arr, target))
```

#6. Move Zeroes at the End

```python
def MoveZeroesAtEnd(nums):
  slow=0
  n=len(nums)
  for fast in range(n):
    if nums[fast]!=0:
      nums[slow],nums[fast]=nums[fast],nums[slow]
      slow+=1

  return nums

nums=[0,1,2,0,3,0,5,0,0,6]
print(MoveZeroesAtEnd(nums))
```

#8. Best Time To Buy and Sell Stock


```python
def StockBuySell(nums):
  max_profit=0
  buy=nums[0]
  n=len(nums)
  for i in range(1,n):
    if buy>nums[i]:
      buy=nums[i]
    else:
      profit=nums[i]-buy
      max_profit=max(profit,max_profit)
  return max_profit

nums = [7,1,5,3,6,4]
print(StockBuySell(nums))
```

#13. Merge Interval

```python
def MergeInterval(intervals):
  intervals.sort(key=lambda x:x[0])

  merge=[]
  for interval in intervals:
    if not merge or merge[-1][1]<interval[0]:
      merge.append(interval)
    else:
      merge[-1][1]=max(merge[-1][1],interval[1])
  return merge

intervals = [[1,3],[2,6],[8,10],[15,18]]
print(MergeInterval(intervals))

intervals = [[1,4],[4,5]]
print(MergeInterval(intervals))
```

#14. Longest Substring Without Repeating Characters

```python
def lengthOfLongestSubstring(s):
  sub_string=set()
  left=0
  max_len=0
  for right in range(len(s)):
    while s[right] in sub_string:
      sub_string.remove(s[left])
      left+=1
    sub_string.add(s[right])

    max_len=max(max_len,right-left+1)

  return max_len

s="abcdab"
print(lengthOfLongestSubstring(s))

s = "abcabcbb"
print(lengthOfLongestSubstring(s))

s = "bbbbb"
print(lengthOfLongestSubstring(s))

s = "pwwkew"
print(lengthOfLongestSubstring(s))
```

#15. Maximum Difference Between Increasing Elements

```python
def maximumDifference(nums):
  max_diff=-1
  min_of_all=nums[0]
  for i in range(len(nums)):
    if nums[i]>min_of_all:
      diff=nums[i]-min_of_all
      max_diff=max(max_diff,diff)
    else:
      min_of_all=nums[i]

  return max_diff

nums=[2,7,3,9,4,5]
print(maximumDifference(nums))

nums = [7,1,5,4]
print(maximumDifference(nums))

nums = [9,4,3,2]
print(maximumDifference(nums))
```

#16. Valid Anagram

```python
def ValidAnagram(s,t):

  if len(t)!=len(s):
    return False

  dict_t={}
  for ele in t:
    if ele in dict_t:
      dict_t[ele]+=1
    else:
      dict_t[ele]=1

  dict_s={}
  for ele in s:
    if ele in dict_s:
      dict_s[ele]+=1
    else:
      dict_s[ele]=1

  if dict_t==dict_s:
    return True
  else:
    return False

s = "anagram"
t = "nagaram"
print(ValidAnagram(s,t))

s = "rat"
t = "car"
print(ValidAnagram(s,t))

s = "rate"
t = "tare"
print(ValidAnagram(s,t))
```

#17. Group Anagrams

```python
def GroupAnagrams(strs):
  result={}
  for ele in strs:
    sorted_strs=''.join(sorted(ele))
    if sorted_strs not in result:
      result[sorted_strs]=[]
    result[sorted_strs].append(ele)
  return result

strs = ["eat","tea","tan","ate","nat","bat"]
print(GroupAnagrams(strs))

strs = [""]
print(GroupAnagrams(strs))
```

#19. Binary Search

```python
def BinarySearch(arr,target):
  left=0
  right=len(arr)-1
  while left<=right:
    mid=(left+right)//2
    if arr[mid]==target:
      return mid
    elif arr[mid]>target:
      right=mid-1
    else:
      left=mid+1
  return -1

nums = [-1,0,3,5,9,12]
target = 9
print(BinarySearch(nums,target))

nums = [-1,0,3,5,9,12]
```

#22. Climb Stair

```python
def ClimbStair(n):
    if n <= 2:
        return n

    dp = [0] * (n + 1)
    dp[1], dp[2] = 1, 2

    for i in range(3, n + 1):
        dp[i] = dp[i - 1] + dp[i - 2]

    return dp[n]

n=6
print(ClimbStair(n))
n=5
print(ClimbStair(n))
n=4
print(ClimbStair(n))
```

#23. Robber House

```python
def RobberHouse(arr):

  dp = [0]*n

  dp[0]=arr[0]
  dp[1]=max(arr[0],arr[1])

  for i in range(2,n):
    dp[i]=max(dp[i-1],dp[i-2]+arr[i])

  return dp,dp[-1]

arr = [1,2,3,1]
print(RobberHouse(arr))

arr = [2,7,9,3,1]
print(RobberHouse(arr))
```

#24. Longest Common Prefix

```python
def LongestCommonPrefix(strs):
  if not strs:
    return ""

  min_len=min(len(ele) for ele in strs)

  for i in range(min_len):
    char=strs[0][i]
    for word in strs:
      if word[i]!=char:
        return strs[0][:i] or 'No Output'
  return strs[0][:min_len] or 'No Output'


strs = ["flower","flow","flight"]
print(LongestCommonPrefix(strs))

strs = ["flower","flow","flowight"]
print(LongestCommonPrefix(strs))

strs = ["dog","racecar","car"]
print(LongestCommonPrefix(strs))
```

#26. Pascal's Triangle

Given an integer numRows, return the first numRows of Pascal's triangle.

In Pascal's triangle, each number is the sum of the two numbers directly above it as shown:

Example 1:

Input: numRows = 5

Output: [[1],[1,1],[1,2,1],[1,3,3,1],[1,4,6,4,1]]

Example 2:

Input: numRows = 1

Output: [[1]]

```python
def generate(numRows):
  result = [[1]]
  for i in range(1, numRows):
    curr = [1]
    k = len(result[-1])
    for j in range(1,k):
      curr_sum = result[-1][j-1] + result[-1][j]
      curr.append(curr_sum)
    curr.append(1)
    result.append(curr)
  return result

numRows = 5
print(generate(numRows))

numRows = 1
print(generate(numRows))

numRows = 3
print(generate(numRows))
```

```python
class ListManager:
    def __init__(self, initial_data=None):
        # O(n) if initial_data is copied, else O(1)
        self.data = initial_data[:] if initial_data else []

    # 1. Insert number at the end
    def insert(self, number):
        # O(1)  append is constant time
        self.data.append(number)

    # 2. Delete the first occurrence of a number
    def delete(self, number):
        # O(n)  list search + shift required
        if number in self.data:
            self.data.remove(number)
        else:
            print(f"{number} not found for delete")

    # 3. Delete all occurrences of a number
    def delete_all(self, number):
        # O(n)  rebuild list without the number
        self.data = [ele for ele in self.data if ele != number]

    # 4. Update first occurrence of old_num to new_num
    def update(self, old_num, new_num):
        # O(n)  search for index and update
        if old_num in self.data:
            index = self.data.index(old_num)
            self.data[index] = new_num
        else:
            print(f"{old_num} not found for update")

    # 5. Check if number exists; if not, append it
    def check_and_add(self, number):
        # O(n) in worst case  contains check
        if number not in self.data:
            self.data.append(number)

    # 6. Get the oldest (first) element
    def get_oldest(self):
        # O(1)  direct access
        return self.data[0] if self.data else None

    # 7. Get the newest (last) element
    def get_newest(self):
        # O(1)  direct access
        return self.data[-1] if self.data else None

    # 8. Add a number (same as insert)
    def add(self, number):
        # O(1)
        self.data.append(number)

    # 9. Check if a number exists in list
    def contains(self, number):
        # O(n)
        return number in self.data

    # 10. Get first index of a number
    def get_index(self, number):
        # O(n)
        if number in self.data:
            return self.data.index(number)
        return -1

    # 11. Count how many times a number occurs
    def count(self, number):
        # O(n)
        return self.data.count(number)

    # 12. Print current list
    def print_list(self):
        print("Current data:", self.data)

    # 13. Insert at specific index
    def insert_at(self, index, value):
        # O(n)  shifting needed
        self.data.insert(index, value)

    # 14. Delete at specific index
    def delete_at(self, index):
        # O(n)  shift after deletion
        if 0 <= index < len(self.data):
            del self.data[index]
        else:
            print("Index out of range")

    # 15. Reverse the list in-place
    def reverse(self):
        # O(n)
        self.data.reverse()

    # 16. Sort the list
    def sort_list(self):
        # O(n log n)  Timsort
        self.data.sort()

    # 17. Clear the list
    def clear(self):
        # O(1)  just clears reference
        self.data.clear()

    # 18. Get element at a specific index
    def get(self, index):
        # O(1)
        if 0 <= index < len(self.data):
            return self.data[index]
        else:
            print("Index out of range")

    # 19. Get sublist (slice)
    def slice(self, start, end):
        # O(k)  k is size of slice
        return self.data[start:end]
```

#26. LinkedList

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.prev = None
        self.next = None

class LinkedListListManager:
    def __init__(self):
        self.head = None
        self.tail = None

    # 1. Insert number at the end - O(1)
    def insert(self, number):
        new_node = Node(number)
        if not self.head:
            self.head = self.tail = new_node
        else:
            self.tail.next = new_node
            new_node.prev = self.tail
            self.tail = new_node

    # 2. Delete the first occurrence of a number - O(n)
    def delete(self, number):
        current = self.head
        while current:
            if current.value == number:
                if current.prev:
                    current.prev.next = current.next
                else:
                    self.head = current.next
                if current.next:
                    current.next.prev = current.prev
                else:
                    self.tail = current.prev
                return
            current = current.next
        print(f"{number} not found for delete")

    # 3. Delete all occurrences of a number - O(n)
    def delete_all(self, number):
        current = self.head
        while current:
            nxt = current.next
            if current.value == number:
                if current.prev:
                    current.prev.next = current.next
                else:
                    self.head = current.next
                if current.next:
                    current.next.prev = current.prev
                else:
                    self.tail = current.prev
            current = nxt

    # 4. Update first occurrence of old_num to new_num - O(n)
    def update(self, old_num, new_num):
        current = self.head
        while current:
            if current.value == old_num:
                current.value = new_num
                return
            current = current.next
        print(f"{old_num} not found for update")

    # 5. Check if number exists; if not, append it - O(n)
    def check_and_add(self, number):
        if not self.contains(number):
            self.insert(number)

    # 6. Get oldest (head) - O(1)
    def get_oldest(self):
        return self.head.value if self.head else None

    # 7. Get newest (tail) - O(1)
    def get_newest(self):
        return self.tail.value if self.tail else None

    # 8. Add number - O(1)
    def add(self, number):
        self.insert(number)

    # 9. Check if number exists - O(n)
    def contains(self, number):
        current = self.head
        while current:
            if current.value == number:
                return True
            current = current.next
        return False

    # 10. Get first index of number - O(n)
    def get_index(self, number):
        current = self.head
        idx = 0
        while current:
            if current.value == number:
                return idx
            current = current.next
            idx += 1
        return -1

    # 11. Count occurrences - O(n)
    def count(self, number):
        current = self.head
        cnt = 0
        while current:
            if current.value == number:
                cnt += 1
            current = current.next
        return cnt

    # 12. Print all elements - O(n)
    def print_list(self):
        current = self.head
        result = []
        while current:
            result.append(current.value)
            current = current.next
        print("Current data:", result)

    # 13. Insert at index - O(n)
    def insert_at(self, index, value):
        new_node = Node(value)
        if index == 0:
            new_node.next = self.head
            if self.head:
                self.head.prev = new_node
            self.head = new_node
            if not self.tail:
                self.tail = new_node
            return
        current = self.head
        for _ in range(index):
            if not current:
                print("Index out of range")
                return
            current = current.next
        if current:
            prev_node = current.prev
            prev_node.next = new_node
            new_node.prev = prev_node
            new_node.next = current
            current.prev = new_node
        else:
            self.insert(value)

    # 14. Delete at index - O(n)
    def delete_at(self, index):
        current = self.head
        for _ in range(index):
            if not current:
                print("Index out of range")
                return
            current = current.next
        if not current:
            print("Index out of range")
            return
        if current.prev:
            current.prev.next = current.next
        else:
            self.head = current.next
        if current.next:
            current.next.prev = current.prev
        else:
            self.tail = current.prev

    # 15. Reverse list - O(n)
    def reverse(self):
        current = self.head
        prev = None
        self.tail = self.head
        while current:
            next_node = current.next
            current.next = prev
            current.prev = next_node
            prev = current
            current = next_node
        self.head = prev

    # 16. Sort list - O(n log n)
    def sort_list(self):
        # convert to list, sort, and rebuild
        values = []
        current = self.head
        while current:
            values.append(current.value)
            current = current.next
        values.sort()
        self.clear()
        for val in values:
            self.insert(val)

    # 17. Clear - O(1)
    def clear(self):
        self.head = None
        self.tail = None

    # 18. Get at index - O(n)
    def get(self, index):
        current = self.head
        for _ in range(index):
            if not current:
                print("Index out of range")
                return
            current = current.next
        return current.value if current else None

    # 19. Slice(start, end) - O(n)
    def slice(self, start, end):
        result = []
        current = self.head
        idx = 0
        while current and idx < end:
            if idx >= start:
                result.append(current.value)
            current = current.next
            idx += 1
        return result
```

#27. HashMap

```python
from collections import defaultdict

class HashMapListManager:
    def __init__(self):
        self.data = []
        self.count_map = defaultdict(int)  # number  count

    # 1. Insert number - O(1)
    def insert(self, number):
        self.data.append(number)
        self.count_map[number] += 1

    # 2. Delete first occurrence - O(n)
    def delete(self, number):
        if number in self.count_map and self.count_map[number] > 0:
            self.data.remove(number)  # O(n)
            self.count_map[number] -= 1
        else:
            print(f"{number} not found for delete")

    # 3. Delete all occurrences - O(n)
    def delete_all(self, number):
        self.data = [x for x in self.data if x != number]
        self.count_map[number] = 0

    # 4. Update old to new - O(n)
    def update(self, old_num, new_num):
        for i in range(len(self.data)):
            if self.data[i] == old_num:
                self.data[i] = new_num
                self.count_map[old_num] -= 1
                self.count_map[new_num] += 1
                return
        print(f"{old_num} not found for update")

    # 5. Check and add - O(1)
    def check_and_add(self, number):
        if self.count_map[number] == 0:
            self.insert(number)

    # 6. Get oldest - O(1)
    def get_oldest(self):
        return self.data[0] if self.data else None

    # 7. Get newest - O(1)
    def get_newest(self):
        return self.data[-1] if self.data else None

    # 8. Add - O(1)
    def add(self, number):
        self.insert(number)

    # 9. Contains - O(1)
    def contains(self, number):
        return self.count_map[number] > 0

    # 10. Get index - O(n)
    def get_index(self, number):
        try:
            return self.data.index(number)
        except ValueError:
            return -1

    # 11. Count - O(1)
    def count(self, number):
        return self.count_map[number]

    # 12. Print list - O(n)
    def print_list(self):
        print("Current data:", self.data)

    # 13. Insert at index - O(n)
    def insert_at(self, index, value):
        self.data.insert(index, value)
        self.count_map[value] += 1

    # 14. Delete at index - O(n)
    def delete_at(self, index):
        if 0 <= index < len(self.data):
            val = self.data.pop(index)
            self.count_map[val] -= 1
        else:
            print("Index out of range")

    # 15. Reverse - O(n)
    def reverse(self):
        self.data.reverse()

    # 16. Sort - O(n log n)
    def sort_list(self):
        self.data.sort()

    # 17. Clear - O(1)
    def clear(self):
        self.data.clear()
        self.count_map.clear()

    # 18. Get at index - O(1)
    def get(self, index):
        if 0 <= index < len(self.data):
            return self.data[index]
        print("Index out of range")

    # 19. Slice - O(k)
    def slice(self, start, end):
        return self.data[start:end]
```
